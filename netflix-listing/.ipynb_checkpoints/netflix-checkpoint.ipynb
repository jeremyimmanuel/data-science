## Loading Data

```python
curr_dir = os.getcwd()
file_path = os.path.join(curr_dir, 'netflix_titles.csv')

df = pd.read_csv(file_path)
df.head()
```

## Selecting Data

Exercising data selection with Pandas

```python
# Get all rows with Brazil as the country

df[df['country'] == 'Brazil']
```

#### Get all Brazillian Movies

Using multi conditional selection to select all rows with Brazil and Movies for its country and type columns. 

A good thing to remember here is we are passing a series to `df`'s square bracket notation. 

```python
brazillian_shows = df['country'] == 'Brazil'
only_movies = df['type'] == 'Movie'

brazillian_shows & only_movies

```

```python
# Get all rows with country Brazil and Movie for type

df[
    (df['country'] == 'Brazil') & (df['type'] == 'Movie')
].head()
```
